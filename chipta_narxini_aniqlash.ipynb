```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python

tr_df = pd.read_csv("train_data.csv")
tr_df
```

```python
x_test = pd.read_csv("test_data.csv")
x_test
```

```python
cat_cols = ['airline', 'source_city', 'departure_time', 'stops',
       'arrival_time', 'destination_city', 'class']
```

```python
def encode_df(dataframe):
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    for column in dataframe.columns:
        if column in cat_cols:
            dataframe[column] = le.fit_transform(dataframe[column])
    return dataframe
```

```python
tr_df = encode_df(tr_df)
x_test = encode_df(x_test)
```

```python
tr_df
```

```python
x_test
```

```python
tr_df.drop(["id","flight"], axis=1, inplace=True)
x_test.drop(["id","flight"], axis=1, inplace=True)
```

```python
dtr = DecisionTreeRegressor(random_state=14)
dtr.fit(tr_df.drop("price", axis=1),y=tr_df['price'])
```

```python
y_pred = dtr.predict(x_test)
y_pred
```

```python
sub = pd.read_csv("sample_solution.csv")
sub
```

```python
sub["price"] = y_pred
sub
```

```python
sub.to_csv("sub-chipta2.csv",index=False)
```

```python
rfr = RandomForestRegressor()
rfr.fit(tr_df.drop("price", axis=1),y=tr_df['price'])
```

```python
y_pred_rfr = rfr.predict(x_test)
y_pred_rfr
```

```python
sub["price"] = y_pred_rfr
sub
```

```python
sub.to_csv("sub-chipta3.csv",index=False)
```
